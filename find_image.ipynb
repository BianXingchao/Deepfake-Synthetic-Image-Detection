```python
import os
import glob
from matplotlib import pyplot
from matplotlib.patches import Rectangle
from matplotlib.patches import Circle
from mtcnn.mtcnn import MTCNN
import skimage.io as io
from PIL import Image
```

```python
count = 0
image_path = []
directory = os.walk(r'/storage0/bxc/data/leaderboard')
for path ,dir_,filelist in directory:
    for filename in filelist:
        if filename.endswith('jpg'):
            image_path.append(path+'/'+filename)
    if len(image_path) == 0:
        continue
image_path.sort()
print("finished.", len(image_path))
print(image_path[:3])
```

```python
fileObject = open(r'/storage0/bxc/data/resize_leaderboard.txt','w')
for path in image_path:
    img = Image.open(path)
    if img.size[0] == 640:
        continue
    else:
        #print(path)
        print(image_path.index(path),path)
        fileObject.write(path)
        fileObject.write('\n')
print('finished resize.')
fileObject.close()
```
